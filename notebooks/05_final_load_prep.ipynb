# 05 Final Load Prep

Use this notebook to compute final KPIs, prepare the Tableau-ready dataset, and export the exact file used in the dashboard.

```python
import pandas as pd

df = pd.read_csv("../data/processed/clean_retail_data.csv")
```

```python
# KPIS
total_revenue = df["total_spent"].sum()
avg_order_value = df["total_spent"].mean()
total_transactions = df.shape[0]

print("Revenue:", total_revenue)
print("Avg Order Value:", avg_order_value)
print("Transactions:", total_transactions)
```

```python
#Category KPI
category_sales = df.groupby("category")["total_spent"].sum().reset_index()
```

```python
#Customer Segmentation
customer_spending = df.groupby("customer_id")["total_spent"].sum()

high_value = customer_spending[customer_spending > customer_spending.quantile(0.75)]
```

```python
df.to_csv("../data/processed/final_ready_data.csv", index=False)
```
